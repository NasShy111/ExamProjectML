```python
#Loading the data#

import pandas as pd

fake = pd.read_csv("../data/Fake.csv")
true = pd.read_csv("../data/True.csv")

print("Fake shape:", fake.shape)
print("True shape:", true.shape)

fake.head()
true.head()

fake["label"] = 0  # fake
true["label"] = 1  # true

df = pd.concat([fake, true], ignore_index=True)

print(df.shape)
df.head()
```

```python
#Cleaning the data
df["text_all"] = (
    df["title"].fillna("") + "\n\n" + df["text"].fillna("")
).str.strip()

#Test
df[["title", "text", "text_all"]].head(2)
```

```python
#Dataset check#

#No of rows
print("Total rows:", df.shape[0])
print(df['label'].value_counts())
print(df['label'].value_counts(normalize=True))  # %

#characters per news
df["n_chars"] = df["text_all"].str.len()
df["n_words"] = df["text_all"].str.split().str.len()

print(df[["n_chars","n_words"]].describe())

#test one fake and one true
df[df["label"]==0].sample(1, random_state=1)[["title","text"]]
df[df["label"]==1].sample(1, random_state=2)[["title","text"]]

#check for sth missing
df.isna().sum()
```
